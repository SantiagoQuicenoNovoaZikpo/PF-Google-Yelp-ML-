### Visualizacion y Analisis con respecto a la columna yelping_since

La columna yelping_since puede mostrarnos en que años la plataforma Yelp tuvo mas registros y en años no

```python
from datetime import datetime as dt
import numpy as np
```

```python
# La funcion convierte cada fila de la columna yelping_since que originalmente es una cadena en formato datetime


def convertir_fecha(fecha) -> str:  

    """La funcion convierte cada ingreso en 'fecha' que originalmente es un string del tipo %Y-%m-%d %H:%M:%S a un datetime con una salida con el formato AAAA-mm-dd
    
        Parametros:
        
        fecha: str"""
    try:
        fecha_t = dt.strptime(fecha, '%Y-%m-%d %H:%M:%S') # convertimos la cadena en datetime en formato AAAA-mm-dd
         
        return fecha_t.date() # retornamos solo el la fecha
    except:
        
        return np.nan # en caso que el campo este vacio o no contenga el formato, returnamos NaN
```

```python
df['yelping_since'] = df['yelping_since'].apply(convertir_fecha) # aplicamos la funcion a cada fila de la columna yelping_since
```

```python
# Visualizamos los datos de yelping_since

counts = df['yelping_since'].value_counts()
plt.plot_date(x=counts.index, y=counts.values, fmt="o")
plt.title('AÑOS CON MAS REGISTROS EN YELP')
plt.xlabel("Fecha")
plt.ylabel("Cantidad de Registros")
plt.show()
```

Los años que la plataforma Yelp tuvo mas ingresos empezaron desde los años 2010 hasta el 2018 aproximadamente
